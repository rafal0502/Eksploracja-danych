```python
# Zaimplementuj algorytm sortowania danej listy przez wybór.
# Algorytm ten w każdej interacji wybiera element najmniejszy na liście
# a nastepnie usuwa go z listy wyjściowej 
# i dodaje go na koniec listy wyjściowej
```

```python
def selection_sort(x):
    x = x.copy()  # tutaj wystarczy kopia płytka
    w = []
    for i in range(len(x)):
        m = min(x)
        x.remove(m)
        w.append(m)
    return w
```

```python
selection_sort([5, 6, 8, 3, 2, 4, 0, 9, 1, 2])
```

```python
# Zaimplementuj algorytm sortowania szybkiego. Algorytm ten polega na 
# podziale elementów listy na mniejsze, równe i większe od wybranego elementu
# zwanego wartownikiem. Następnie rekurencyjnie wywołujemy ten sam algorytm 
# dla pierwszego i trzeciego podciągu i zwracamy listę będącą złączeniem tak
# skonstruowanych trzech podciągów (drugiego nie trzeba sortować)
```

```python
def quick_sort(x):
    if len(x) <=1: 
        return x
    v = x[0]   #wartownik
    l = quick_sort([a for a in x if a < v])
    m =            [a for a in x if a == v]
    r = quick_sort([a for a in x if a > v])
    return l + m + r


print(quick_sort([5, 6, 8, 3, 2, 4, 0, 9, 1, 2]))
print(quick_sort("mielonka"))
```

```python
# Niech będzie dany obiekt iterowalny złożony z obiektów typu niezmienialnego,
# np. lista liczb całkowitych. Wyznacz modę (dominantę), zcyli najczęściej
# występującą w nim wartość

def moda(x):
    # wyznacz liczbę wystąpień każdego elementu
    s = dict() # pusty słownik
    for e in x:
        if e in s:
            s[e] += 1
        else: 
            s[e] = 1
    # znajdź element o największej wartości
    ile, co = -1, None
    for co2, ile2 in s.items():
        if ile2 > ile:
            ile, co = ile2, co2
    return co # zwróć odpowiadający mu klucz
```

```python
moda([1, 2, 3, 1, 2, 4, 3, 6, 5, 3, 6, 3])
```

```python
def ile_poczatkowych(x, w):
    suma = 0
    for i in range(len(x)):
        suma += x[i]
        if suma >= w:
            break #mamy naszą sumę więc przerywamy
    return i + 1
```

```python
ile_poczatkowych([1, 2, 3, 4, 5], 4)
```

```python
def ile_poczatkowych2(x, w):
    suma = 0
    for i in range(len(x)):
        if x[i] % 2 == 1: continue
        suma += x[i]
        if suma >= w:
            break
    else:
        raise Exception("To jest ex-papuga!")
    return i + 1
```

```python
# czyli czy dzieli się tylko przez samą siebie oraz 1
```

```python
def pierwsza(n):
    if n<=1:
        return False # z definicji
    k = 2
    while k <=n ** 0.5: #wystarczy pierwiastek(n)
        if n%k == 0 :
            return False
        else:
            k+=1
    return True
```

```python
for i in range(1,6):
    print(i, pierwsza(i))
```

```python
# Funkcja wyliczająca max ,ale bez używania max()
```

```python
def gdzie_maks(x):
    it = enumerate(x)
    indeks, maks = next(it)
    for i, e in it:
        if e > maks:
            indeks, maks = i, e
    return indeks

```

```python
gdzie_maks([1, 2, 3, 4, 5, 6, 4, 5, 6, 3, 2, 1])
```

```python
gdzie_maks("mielonka")
```

```python
gdzie_maks(reversed("mielonka"))
```

```python
# Napisz funkcję minmaks(), która dla iterowalnego obiektu zwraca dwuelementową krotkę zawierającą
# jego element najmniejszy i największy. Użyj tylko jednej pętli for.
```

```python
def minmaks(x):
    it = iter(x)
    min = max = next(it)
    for e in it:
        if e < min:
            min = e
        elif e > max:
            max = e
    return min, max
```

```python
minmaks([1,4,7,3,9,6,0,9,3])
```

```python
def maks3(a,b,c):
    "Wyznacza maksimum z 3 podanych argumentów."
    if b <= a >= c: return a
    if a <= b >= c: return b
    else:           return c
```

```python
maks3(3,1,2)
```

```python
def fib1(n):
    "Wyznacza n-tą liczbę Fibonacciego (rekurencyjnie)."
    if n <= 1: # warunek stopu
        return 1
    else:
        return fib1(n-1) + fib1(n-2)
        
```

```python
def fib2(n):
    "Wyznacz n-tą liczbą Fibonacciego (iteractjnie)."
    a, b = 1 , 0
    for i in range(n):
        a, b = a+ b, a
    return a
```

```python
def fib3(n):
    "Wyznacz n-tą liczbą Fibonacciego (iteractjnie)."
    assert isinstance(n,int), "n nie jest liczba całkowitą"
    assert n >= 0, "n nie jest nieujemne"
    a, b = 1 , 0
    for i in range(n):
        a, b = a+ b, a
    return a
```

```python
white_wine = wine[wine.color == "white"]
```

```python
white_wine = white_wine.iloc[:, 0:11]
```

```python
# interesuje nas wyznaczenie modelu regresji dla kolumny alcohol
```

```python
y = white_wine.iloc[:, -1] # alcohol
```

```python
X = white_wine.iloc[:, :-1] # ...to ostatnia kolumna
```

```python
# chcemy się dowiedzieć, czyu alkohol jest funkcją pozostałych 10 zmiennych i jeśli tak, jakiej ta funkcja jest postaci
```

```python
# wyznaczamy wartości r czyli korelacji liniowej Pearsona : jeśli jest bliska 1 lub -1 to istnieje silna, odpowiednio dodatnia lub
# ujemna zależnośc liniowa miedzy para zmiennych
```

```python
# wyznaczamy wartosci r dla wszystkich par zmiennych ilosciowych w white_wine:
c = white_wine.corr("pearson")
```

```python
c
```

```python
# wykaz par skorelowanych ze soba najsilniej:
c = c.where (
        np.triu( # macierz trójkatna górna
                 # -> tylko unikatowe pary zmiennych
                 np.ones(c.shape,
                         dtype = np.bool),
                         k = 1)
               ).stack().sort_values()
```

```python
c[abs(c)>0.5]
```

```python
import sklearn.linear_model
```

```python
mnk = sklearn.linear_model.LinearRegression()
```

```python
# aby wyznaczyc parametr modelu na podstawie podanej proby uczacej wywolujemy metode fit
```

```python
mnk.fit(X,y) # dopasowanie modelu do danych
```

```python
# oszasowanie wspolczynnikow beta w metodzie najmniejszych kwadratow odczytujemy:
```

```python
mnk.intercept_ # wyraz wolny, beta0
```

```python
mnk.coef_ # dalsze wspolczynniki
```

```python
# na podstawie dopasowanego modelu mozemy dokonac prognozowania ilosci alkoholu. Dla (juz teraz dowolnego)
# x mozemy wyznaczyc wartosci wywolujac metode predict():
```

```python
x_nowy = X.mean().reshape(1, -1) # średnie wino
mnk.predict(x_nowy)
```

```python
# Ocena jakosci dopasowania modelu
```

```python
y_pred = mnk.predict(X)
```

```python
mnk.score(X, y)
```
